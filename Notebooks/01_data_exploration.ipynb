## 📥 1. Imports
We import essential libraries for data manipulation, visualization, and formatting.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## 📂 2. Load Dataset
Load the CSV file containing the electricity consumption data.

```python
file_path = r'C:\Users\Bahae\electricity_consumption_forecasting\Datasets\powerconsumption.csv'  # replace with your file path
df = pd.read_csv(file_path)
```

## 🕓 3. Convert Datetime
Convert the 'Datetime' column to a proper datetime format for time series analysis.

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
```

## 🧾 4. Basic Information
Display general information about the dataset, including data types and non-null counts.

```python
print(df.info())
```

## 🧹 5. Check for Missing Values
Identify any missing values in the dataset.

```python
print("\nMissing values:")
print(df.isnull().sum())
```

## 👀 6. Preview the Data
View the first few rows of the dataset to understand its structure.

```python
print("\nFirst 5 rows:")
df.head()
```

## ⏳ 7. Set Datetime as Index
Make 'Datetime' the index to enable time-based operations.

```python
df.set_index('Datetime', inplace=True)
```

## 📊 8. Plot Power Consumption
Visualize electricity consumption across all three zones over time.

```python
plt.figure(figsize=(14, 6))
df['PowerConsumption_Zone1'].plot(label='Zone 1')
df['PowerConsumption_Zone2'].plot(label='Zone 2')
df['PowerConsumption_Zone3'].plot(label='Zone 3')
plt.title("Power Consumption Over Time")
plt.ylabel("Power (Watts)")
plt.legend()
plt.show()
```

## 🧠 9. Correlation Heatmap
Check how numerical variables relate to each other using a heatmap.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
